```python
%%manim -qm -v WARNING Capture

class Capture(Scene):
    
    def construct(self):
        self.anim_capture()

    def anim_capture(self):
        """
        Animation permettant de visualiser le role de capture dans le code
        
        Que fait la fonction capture ? 
        
        1 - essaie d'appliquer le code contenu après 
            Si fonctionne, l'execute
            Si ne fonctionne pas, ne fait rien
            
        Comment la représenter ? 
        
        1 - Créer un rectangle
        2 - Ecrire un texte dans le rectangle, qui représente la fonction
        3 - Afficher un message temporaire, en jaune, type "executer"
        4 - Afficher une tick verte 
        5 - Décaler vers la gauche la boite
        6 - Afficher : la fonction fonctionne
        7 - A droite, afficher un texte "resultat de la fonction"
        8 - Tout effacer
        9 - Ramener la boite au centre 
        10 - Afficher une tick rouge
        11 - Afficher un texte genre "erreur"
        12 - Decaler sur la gauche
        13 - Afficher en rouge un message d'erreur
        
        """
        
        fonction=VGroup()
        
        #1 - Créer un rectangle appelé capture
        rect = Rectangle(width=5)
        self.add(rect)
        fonction.add(rect)
        
        capture = Text("Capture")
        capture.to_corner(UL)
        self.add(capture)
        fonction.add(capture)
        
        
        #2 - Ecrire un texte dans le rectangle, qui représente la fonction à capturer
        nom_fonction = Text("Une fonction")
        #nom_fonction.next_to(rect, UP)
        self.play(FadeIn(nom_fonction))
        self.wait()
        
        nom_fonction_reel = Code(
            code="program drop figure", 
            language="stata",
        
        )
        
        nom_fonction_reel.next_to(rect, UP)
        
        self.play(Transform(nom_fonction, nom_fonction_reel))
        self.wait()
        
        
        #3 - Afficher un message temporaire, en jaune, type "executer"
        
        
        #4 - Afficher une tick verte 
        
        self.execution(rect, True)
        self.wait()
        
        self.execution(rect, False)
        
        #11 - Afficher un texte genre "erreur"
        #12 - Decaler sur la gauche
        #13 - Afficher en rouge un message d'erreur
        self.wait()
        
    
    def execution(self, rect, reussite=True):
        
        antcd = Dot()
        antcd.next_to(rect, LEFT)
        self.play(FadeIn(antcd))
        self.wait()
        
        self.play(antcd.animate.shift(RIGHT * 3))
        
        execute = Text("Execution", color=YELLOW)
        execute.next_to(rect, DOWN)
        self.play(FadeIn(execute))
        self.wait()
        self.remove(execute)

        retour = Code(
            code="program drop figure", 
            language="stata",
        )

        retour.next_to(rect, RIGHT*2)
        commentaire_return = Text("La fonction est exécutée")
        commentaire_return.next_to(antcd, UP)

        commentaire_echec = Text("Rien ne se passe : la ligne est ignorée")

        if reussite:
            succes = Text("Succès", color=GREEN)
        else:
            succes = Text("Echec", color=RED)

        succes.next_to(rect, DOWN)
        self.add(succes)
        self.wait()



        #5 - Décaler vers la gauche la boite
        #6 - Afficher : la fonction fonctionne

        self.play(antcd.animate.shift(RIGHT*5))


        #7 - A droite, afficher un texte "resultat de la fonction"

        if reussite:
            self.play(ReplacementTransform(antcd, retour))
            commentaire_return.next_to(succes, DOWN)
            self.play(FadeIn(commentaire_return))
            self.wait(3)
            self.remove(commentaire_return)
            self.play(retour.animate.shift(DOWN*10))
        else:
            self.wait(1.5)
            commentaire_echec.next_to(succes, DOWN)
            self.add(commentaire_echec)
            self.play(FadeOut(antcd))
            
        self.remove(succes)
        self.remove(retour)

        #8 - Tout effacer
        #9 - Ramener la boite au centre 
        #10 - Afficher une tick rouge
```

```python
%%manim -qm -v WARNING Program

class Program(Scene):
    
    def construct(self):
        self.anim_program()
    
    def anim_program(self):
        """
        Animation permettant de comprendre une fonction en stata
        
        On peut reprendre une partie du code ci dessus
        
        1 creer un rectangle
        2 entrer une forme dans le rectangle
        3 transformer cette forme dans le rectangle
        4 sortir cette forme du rectangle
        5 mettre des lignes de code dans le rectangle
        
        """
        
        
        
        rect = Rectangle(width=6, height=4)
        self.add(rect)
        
        legende = Text("Fonction")
        legende.next_to(rect, DOWN)
        self.play(FadeIn(legende))
        
        text_input = Text("input", color=RED)
        text_output = Text("output",color=BLUE)
        
        
        
        cercle = Circle()
        cercle.next_to(rect, LEFT)
        self.play(FadeIn(cercle))
        
        text_input.next_to(cercle, DOWN)
        self.play(FadeIn(text_input))
        self.wait()
        self.play(FadeOut(text_input))
        
        
        
        triangle = Triangle()
        
        
        
        self.play(cercle.animate.shift(RIGHT*4))
        self.wait()
        
        self.play(ReplacementTransform(cercle, triangle))
        self.play(triangle.animate.shift(RIGHT*5))
        
        text_output.next_to(triangle, DOWN)
        self.play(FadeIn(text_output))
        self.wait()
        self.play(FadeOut(text_output, triangle))
        
        program = Code(
            code="program ", 
            language="stata",
        )
                
        program_define = Code(
            code="program define", 
            language="stata",
        )
        
        program_define_figure = Code(
            code="program define figure", 
            language="stata",
        )
        
        
        program.next_to(rect, UP)
        program_define.next_to(rect, UP)
        program_define_figure.next_to(rect, UP)
        
        text_program = Text("Commande pour les programmes")
        text_program.next_to(rect, DOWN)
        
        text_define = Text("Indique qu'on va définir un programme")
        text_define.next_to(rect, DOWN)
        
        text_figure = Text("Nom du programme")
        text_figure.next_to(rect, DOWN)
        
        self.play(FadeOut(legende))
        self.play(FadeIn(program, text_program))
        self.wait()
        
        self.play(Transform(program, program_define))
        self.play(Transform(text_program, text_define))
        self.wait()
        
        
        self.play(Transform(program, program_define_figure))
        self.play(Transform(text_program, text_figure))
        self.wait()
        
        self.play(FadeOut(program, text_program))
        
        self.add(program_define_figure)
        
        self.wait()
        
        texte_programme = MathTex("program"," ", "define"," ","figure")
        texte_programme.next_to(rect, UP)
        
        self.play(Transform(program_define_figure, texte_programme))
        
        framebox2 = SurroundingRectangle(texte_programme[0], buff = .1)
        
        self.play(Create(framebox2))
```

```python
%%manim -qm -v WARNING Bandwidth

class Bandwidth(Scene):
    
    def construct(self):
        self.anim_program()
    
    def anim_program(self):
        """
        Animation permettant de comprendre la notion de bandwidth 
        
        1. Creer un graphique
        2. On affiche en abscisse povrate60
        3. On définit une droite de valeur x = povrate
        4. On créé une bande colorée qui part de cette droite et qui s'étend à +8, -8
        5. On la transforme en un carré violet
        6. On la transforme en le mot bandwidth en violet
        
        """
        
        graphe = NumberPlane(
            x_range=(0,10,1),
            y_range=(0,6, 1), 
            background_line_style={
                "stroke_color": BLUE,    # Couleur des lignes de quadrillage
                "stroke_width": 2,        # Épaisseur des lignes de quadrillage
                "stroke_opacity": 0.5,    # Opacité des lignes de quadrillage
            }
        )
        
        self.add(graphe)
        
        self.wait()
        
        legende = Text("povrate60")
        legende.next_to(graphe, DR)
        
        self.play(FadeIn(legende))
        
        ligne = Line([0, -3, 0],[0, 3, 0], color=RED)
        self.play(Create(ligne))
        
        self.wait(2)
        
        bandwidth = Rectangle(width=3.5, height=6, color=PURPLE, fill_color=PURPLE, fill_opacity=1)
        self.play(Create(bandwidth))
        #self.play(Transform(ligne, bandwidth))
        
        carre = Square(1,color=PURPLE)
        carre.next_to([2,2,0])
        
        self.play(Transform(bandwidth, carre))
        
        self.wait(3)
        
        texte = Text("Bandwidth", color=PURPLE)
        texte.next_to(carre, DOWN)
        
        self.play(ReplacementTransform(bandwidth, texte))
        
        self.wait(3)
    
```
